```python
import pandas as pd
import numpy as np
```

Импортируйте из соответствующих модулей RandomForestClassifier, GridSearchCV и train_test_split.

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import train_test_split
```

Загрузите датасет creditcard.csv и создайте датафрейм df.

```python
DATASET_PATH = './creditcard.csv'
PREPARED_DATASET_PATH = './creditcard_prepared.csv'

df = pd.read_csv(DATASET_PATH, sep=',')
df.head(3)
```

Создайте датафрейм X из датафрейма df, исключив столбец Class.

```python
target = "Class"
X = df.drop(target, axis=1)
X.head(5)
```

Создайте объект Series под названием y из столбца Class.

```python
y = df[target]
```

Разбейте X и y на тренировочный и тестовый наборы данных при помощи функции train_test_split, используя аргументы: test_size=0.3, random_state=100, stratify=y.

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=100, stratify=y)
```

У вас должны получиться объекты X_train, X_test, y_train и y_test.

Для поиска по сетке задайте параметры

```python
parameters = [{'n_estimators': [10, 15],
'max_features': np.arange(3, 5),
'max_depth': np.arange(4, 7)}]
```

Создайте модель GridSearchCV

```python
modelGS = GridSearchCV(
    estimator=RandomForestClassifier(random_state=100),
    param_grid=parameters,
    scoring='roc_auc',
    cv=3
)
```

Обучите модель на тренировочном наборе данных (может занять несколько минут).

```python
modelGS.fit(X_train, y_train)
```

Просмотрите параметры лучшей модели с помощью атрибута best_params_

```python
modelGS.best_params_
```

Предскажите вероятности классов с помощью полученнной модели и метода predict_proba.

```python
y_pred_proba = modelGS.predict_proba(X_test)
print(y_pred_proba[:10])
```

Из полученного результата (массив Numpy) выберите столбец с индексом 1 (вероятность класса 1) и запишите в массив y_pred_proba. Из модуля sklearn.metrics импортируйте метрику roc_auc_score.

```python
y_pred_proba = y_pred_proba[:, 1]
print(y_pred_proba[:5])
```

```python
from sklearn.metrics import roc_auc_score
```

Вычислите AUC на тестовых данных и сравните с результатом,полученным на тренировочных данных, используя в качестве аргументов массивы y_test и y_pred_proba.

```python
roc_auc_score(y_test, y_pred_proba)
```
